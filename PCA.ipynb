```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
Properties_Dataset=pd.read_excel('PCADataset.xlsx')
Properties_Dataset.head()
```

```python
#DATA ANALYSIS AND FEATURE ENGINEERING
```

```python
Properties_Dataset.shape
```

```python
Properties_Dataset=Properties_Dataset.drop(columns=['Catalyst','Catalyst E1','Catalyst E2','Catalyst E3','Catalyst E4'],axis=1)
Properties_Dataset.head()
```

```python
Properties_Dataset.isnull().sum()
```

```python
#ELEMENT 1
```

```python
DataE1=Properties_Dataset.iloc[:,0:20]
DataE1.head()
```

```python
DataE1.isnull().sum()
```

```python
DataE1.corr()
```

```python
Standardize=StandardScaler()
DataE1=Standardize.fit_transform(DataE1)
DataE1.shape
```

```python
estimator=PCA(n_components=4)
DataE1=estimator.fit_transform(DataE1)
DataE1.shape
```

```python
print(estimator.explained_variance_ratio_)
```

```python
DataE1=pd.DataFrame(DataE1)
DataE1.head()
```
